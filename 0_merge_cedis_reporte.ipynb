# Carga y preparación de [df]


```python
day = "0129" # SE MODIFICA DIARIO
```

```python
# DataFrame
# "C:\Users\FernandoManuelGamino\Documents\4_cierre_dia_coordinadores\prueba\cedis.xlsx"

files = glob.glob("../4_cierre_dia_coordinadores/" + day + "/*.xlsx")
df_diario = []
```

```python
for file in files:
    temp_df = pd.read_excel(file, sheet_name = "RPT_Rejection_Rabbit")
    df_diario.append(temp_df)
    nacional = pd.concat(df_diario) 
    print(f'Successfully created dataframe for {file} with shape {temp_df.shape}')
    print('and added it to macro_df')
```
