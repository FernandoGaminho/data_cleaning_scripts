```python
import pandas as pd
import glob
import re
import os.path
from datetime import datetime
```

# Carga y preparación de [df]


```python
week = "01" # <<<<<<< MODIFICAR SEMANALMENTE
```

```python
files = glob.glob("../historico_operaciones/*.xlsx")
df_diario = []
```

```python
for file in files:
    temp_df = pd.read_excel(file)
    df_diario.append(temp_df)
    df = pd.concat(df_diario) 
    print(f'Successfully created dataframe for {file} with shape {temp_df.shape}')
    print('and added it to macro_df')
```

# Exportación de archivo

```python
file = week+".xlsx"
joined_path = os.path.join("../historico_operaciones/", file)
df.to_excel(joined_path,sheet_name=week, index=False)
```
